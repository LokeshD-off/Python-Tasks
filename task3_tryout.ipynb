```python
import csv
import requests
from bs4 import BeautifulSoup

##obtaining the extracted text without trailing whitespaces
def extract_text_from_url(url):
    res = requests.get(url)
    soup = BeautifulSoup(res.content,'html.parser')
    text = soup.get_text()
    return text.strip()

##creating an output file to write down the text 
with open('output_task3.csv','w',newline = '') as writefile:
    writing = csv.writer(writefile)
    with open('DMining_Pandas(Sheet1).csv','r',encoding = 'latin-1') as readfile:
        reading = csv.reader(readfile)
        next(reading)
        for row in reading:
            individual_url = row[2]
            text_from_each_pdf = extract_text_from_url(individual_url)
            writing.writerow(text_from_each_pdf)
```

```python
##create an output proxy list to store each 1024 characters 
def processing_each_chunk(chunk,id):
    cleaned_chunk = chunk.strip()
    cleaned_chunk = cleaned_chunk.replace("\n"," ")
    print(len(cleaned_chunk))
    return {'id': id, 'chunk': cleaned_chunk}
chunk_size = 1024
processed_chunks = []

for i in range(0,len(text),chunk_size):
    chunk = text[i:i+chunk_size]
    processed_chunk = processing_each_chunk(chunk, f'chunk_{i//chunk_size}')
    processed_chunks.append(processed_chunk)

csv_filename = "output_test_task3.csv"
with open(csv_filename,'w') as writefile_3:
    fieldnames = ['id','chunk']
    writing_3 = csv.DictWriter(writefile_3, fieldnames=fieldnames) 

    writing_3.writeheader()
    for chunk in processed_chunks:
        writing_3.writerow(chunk)

print(f"Processed chunks saved to {csv_filename}.")
```

```python
##culmination of all different steps into one to extract data from one pdf and put it into the output as required
import csv
import requests
import fitz

#Locally download all the files
with open("DMining_Pandas(Sheet1).csv",'r',encoding= 'latin-1') as datafile:
    reading_3 = csv.reader(datafile)
    next(reading_3)
    i = 1
    for data in reading_3:
        tbdownloaded_url = data[2]
        response_3 = requests.get(tbdownloaded_url)

        if response_3.status_code == 200:
            with open("test_download_file_{}.pdf".format(i),'wb') as writefile_4:
                writefile_4.write(response_3.content)
            print("File Downloaded successfully!")
        i = i+1

#cleaning chunk and returning key-value pair
def process_each_chunk(chunk,id):
    clean_chunk = chunk.strip()
    clean_chunk = clean_chunk.replace("\n"," ")
    return {'id': id, 'chunk' : clean_chunk}

#Extract text from the downloaded files and process each chunk and store it in processed_chunks_2 list
chunk_size = 1024
processed_chunks_2 = []
for j in range(1,i):
    doc_2 = fitz.open("test_download_file_{}.pdf".format(j))
    text = ""

    for page in doc_2:
        text += page.get_text()
    print("one file has been put into the text")

    for k in range(0,len(text),chunk_size):
        chunk = text[k:k+chunk_size]
        processed_chunk_2 = process_each_chunk(chunk,f'chunk_{k}_test_download_file{j}')
        processed_chunks_2.append(processed_chunk_2)

##Creating an csv file to store the processed_chunks_2 list as Dictionary 
output_file =  "output_finaltest_task3.csv"
with open(output_file,'w') as opfile:
    fieldnames_2 = ['id','chunk']
    writing_4 = csv.DictWriter(opfile,fieldnames=fieldnames_2)

    writing_4.writeheader()
    for singlerow in processed_chunks_2:
        writing_4.writerow(singlerow)
print("Complete output has been imprinted on to {}".format(output_file))
    
```
